```python

# Iterate over rows and remove rows with null values in column 'A'
rows_to_keep = []

for index, row in df.iterrows():
    if pd.notnull(row['Cast']):  # Check if the value in column 'A' is not null
        rows_to_keep.append(index)  # If not null, keep this row

# Create a new DataFrame with only the rows that don't have null values in column 'A'
df = df_cleaned = df.loc[rows_to_keep]

# Display the updated DataFrame
print(df_cleaned)
```
